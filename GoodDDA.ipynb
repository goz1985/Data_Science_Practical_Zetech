```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = max(abs(dx), abs(dy))
    x_increment = dx / steps if steps != 0 else 0
    y_increment = dy / steps if steps != 0 else 0

    x, y = x1, y1
    points = []

    for _ in range(int(steps) + 1):
        points.append((round(x), round(y)))
        x += x_increment
        y += y_increment

    return points

# Example usage:
x1, y1 = 2, 2
x2, y2 = 30, 18

result = dda(x1, y1, x2, y2)
print(result)

# Plotting the line using matplotlib
x_values, y_values = zip(*result)
plt.plot(x_values, y_values, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('DDA Line Drawing')
plt.grid()
plt.show()
51
```
